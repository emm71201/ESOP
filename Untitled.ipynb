```python
def get_anfs(file):
    anfs = {}
    count = 0
    with open(file, "r") as infile:
        tmp = ""
        for line in infile:
            if line.strip() == "END":
                anfs[count] = tmp.split(",")[1]
                tmp = ""
                count += 1
            else:
                tmp += line.strip()
    
    return anfs

anfs = get_anfs("s108_anfs.txt")
```

```python
def parse_anf(n, anf):
    
    anf = anf.replace("\\", "").replace("(", "").replace(")", "").split("^")
    
    cubes = []
    
    for cube in anf:
        
        expr = ["-"]*n
        for literal in cube:
            if literal != "&":
                indx = n - alphabet.index(literal) - 1
                expr[indx] = "1"
        
        cubes.append(Cube("".join(expr)))
    
    return ESOP(cubes)

```

```python
n = 16
esop = parse_anf(n, anfs[0])
```

```python
esops = {}
for key,item in anfs.items():
    esop = parse_anf(n, item)
    esop_reduced = ESOP(esop.reduce())
    esops[key] = esop_reduced
```

### Quantum Circuit

```python
from qiskit import QuantumCircuit, QuantumRegister
```

```python
t = QuantumRegister(1, "t")
s = QuantumRegister(1, "s")
r1 = QuantumRegister(1, "r_1")
r0 = QuantumRegister(1, "r_0")
q1 = QuantumRegister(1, "q1")
q0 = QuantumRegister(1, "q0")
p1 = QuantumRegister(1, "p_1")
p0 = QuantumRegister(1, "p_0")


tp = QuantumRegister(1, "t'")
sp = QuantumRegister(1, "s'")
rp1 = QuantumRegister(1, "r_1'")
rp0 = QuantumRegister(1, "r_0'")
qp1 = QuantumRegister(1, "q_1'")
qp0 = QuantumRegister(1, "q_0'")
pp1 = QuantumRegister(1, "p_1'")
pp0 = QuantumRegister(1, "p_0'")

anc = QuantumRegister(1, "anc")
qc = QuantumCircuit(t, s, r1, r0, q1, q0, p1, p0,tp, sp, rp1, rp0, qp1, qp0, pp1, pp0, anc)
qc.reset(anc)
print("setup done")
```

```python
# compute t''
qc.cx(t,tp)
qc.barrier()
qc.draw()

# compute s''
qc.cx(s,sp)
qc.barrier()
qc.draw("mpl")
```
